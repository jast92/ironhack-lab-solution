## Check and deal with NaN values.

```python
nan_value = 0
for i in data.columns.tolist():
    if data[i].isnull().values.any() == True:
        nan_value += 1
if nan_value == 0:
    print("There are no missing values")
else:
    print("There are {%} columns with at least one missing value" %nan_value)    
```
